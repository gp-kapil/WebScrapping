```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options
import math
```

```python
url = "https://haryanarera.gov.in/admincontrol/registered_projects/2"
driver = webdriver.Firefox()
driver.get(url)
```

```python
Elements_by_xpath = []
Elements_by_xpath.append('//*[@id="compliant_hearing_paginate"]/span/a[2]')
Elements_by_xpath.append('//*[@id="compliant_hearing_paginate"]/span/a[3]')
Elements_by_xpath.append('//*[@id="compliant_hearing_paginate"]/span/a[4]')
Elements_by_xpath.append('//*[@id="compliant_hearing_paginate"]/span/a[5]')
```

```python
Project_Name = []
Builder =[]
Project_Location = []
Project_District = []
Registered_With = []
Project_Detail_URL = []
```

```python
def extract_data():
    
    soup = BeautifulSoup(driver.page_source, "html.parser")
    table = soup.find_all('table',attrs={'id':'compliant_hearing'})
    table_body = table[0].find_all('tbody')
    table_row = table_body[0].find_all('tr',attrs={'role':'row'})
    
    for row in table_row:
        
        table_div = row.find_all('td')
    
        Project_Name.append(table_div[3].text)
        Builder.append(table_div[4].text)
        Project_Location.append(table_div[5].text)
        Project_District.append(table_div[6].text)
        Registered_With.append(table_div[7].text)
        if(table_div[8].find('a')):
            Project_Detail_URL.append(table_div[8].find('a')['href'])
        else:
            Project_Detail_URL.append(None)

    
    
```

```python
extract_data()
for xpath in Elements_by_xpath:
    driver.find_element_by_xpath(xpath).click()
    extract_data()
```

```python
Dataset = pd.DataFrame({'Project Name':Project_Name,
                           'Builder':Builder,
                           'Project Location':Project_Location,
                           'Project District':Project_District,
                           'Registered With':Registered_With,
                           'Project Detail URL':Project_Detail_URL})
```

```python
Dataset.to_csv('RERA_HARAYANA_LEADS.csv',index=False)
```
