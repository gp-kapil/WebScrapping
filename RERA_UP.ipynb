```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options
import math
import os
```

```python
def Extract_Data():
    
      
    soup = BeautifulSoup(driver.page_source, "html.parser")
    rows = soup.find("table", border=0).find("tbody").find_all("tr")
       
    column_names = ["ProjectName", "Promoter", "District","ProjectType","ID"]
    df = pd.DataFrame(columns = column_names,index = range(len(rows)))

    counter = 0
    for row in rows:
        if(row.find('td')):
            cells = row.find_all("td")
            #print(cells)
            print(cells[2].text) #Project Name
            print(cells[3].find('li').text) # Promoter
            print(cells[4].text) #District
            print(cells[5].text) #Project Type
            print(cells[7].find('a').get('id')) # Next Page Button ID
            print("*"*50)
            
            df.loc[counter].ProjectName = cells[2].text.strip()
            df.loc[counter].Promoter = cells[3].find('li').text.strip()
            df.loc[counter].District = cells[4].text.strip()
            df.loc[counter].ProjectType = cells[5].text.strip()
            df.loc[counter].ID = cells[7].find('a').get('id').strip()
            
            counter += 1
            
        else:
            continue
    
    return df
```
