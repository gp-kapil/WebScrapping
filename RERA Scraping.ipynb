```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options
```

```python
District_l = []
Project_l = []
Type_l = []
Promoter_l = []
Application_l = []
Registration_l =[]
url_l =[]
```

```python
url = "http://rera.rajasthan.gov.in/ProjectSearch"
driver = webdriver.Firefox()
driver.get(url)
soup = BeautifulSoup(driver.page_source, "html.parser")
result = soup.find('div',attrs={'id':'dvGrid','style':True})

#Check If project List Table is Visible or not
if(result['style'] == 'display:none;'):
    search_button = driver.find_element_by_xpath('//*[@id="btn_SearchProjectSubmit"]')
    search_button.click()
    time.sleep(4)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    list_pager = soup.find_all('span',class_="ds4u-btn-content")
    #print(list_pager)
    #print(list_pager[-1].text)

    #Paginate Tables 
    for i in range(1,int(list_pager[-1].text)+1):
        #driver.findElement(By.linkText(str(i))).sendKeys(Keys.ENTER);
        #print(i)
        driver.find_element_by_link_text(str(i)).click()
        soup = BeautifulSoup(driver.page_source, "html.parser")
        
        #Get all visible Table Row
        result = soup.find_all('tr',class_="ds4u-row")
        
        #Get all column value within a row
        for row in result:
            
            columns = row.find_all('td')
            counter = 0;
            
            for col in columns :
                
                content = ""
                if(col.find('a',class_="btn btn-sm btn-primary")):
                    anchor = col.find('a',class_="btn btn-sm btn-primary")
                    content = anchor['href']
                else:
                    content = col.text
                
                #Insert Data into list
                list_to_insert = [District_l,
                                  Project_l,
                                  Type_l,
                                  Promoter_l,
                                  Application_l,
                                  Registration_l,
                                  url_l]
                list_to_insert[counter].append(content)
                counter += 1
        #print(url_l)
        
        time.sleep(10)

#Quit Selenium
driver.quit()
```

```python
Dataset = pd.DataFrame({'District':District_l,
                           'Project':Project_l,
                           'Type':Type_l,
                           'Promoter':Promoter_l,
                           'Application no.':Application_l,
                           'Registration no.':Registration_l,
                           'project url':url_l})
```

```python
anchor_url = 'http://rera.rajasthan.gov.in'

Project_Name_list = []
Project_Type_list = []

Promoter_Name_list = []
Promoter_Address_list = []
Promoter_District_list = []
Promoter_Contact_Number_list = []
Promoter_Website_list = []

Project_Commencement_Date_list = []
Project_Completion_Date_list = []
Project_Tehsil_list = []
Project_Plot_list = []
Project_Street_list =[]
Project_Town_list = []

Project_Number_of_Apartments_list = []
Project_Number_of_Sanctioned_Apartment_list = []

Project_Contractor_Name_list = []
Project_Contractor_Email_Address_list = []
Project_Contractor_Contact_Address_list = []
Project_Contractor_Contact_Number_list = []

Project_Architect_Name_list = []
Project_Architect_Email_Address_list = []
Project_Architect_Contact_Address_list = []
Project_Architect_Contact_Number_list = []

Project_Structural_Engineer_Name_list = []
Project_Structural_Engineer_Email_Address_list = []
Project_Structural_Engineer_Contact_Address_list = []
Project_Structural_Engineer_Contact_Number_list = []



for i in range (554, Dataset.shape[0]):
    extension = Dataset.iloc[i]['project url']
    URL = anchor_url+extension
    
    driver = webdriver.Firefox()
    driver.get(URL)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    
    Project_Name_list.append(Dataset.iloc[i]['Project'])
    Project_Type_list.append(Dataset.iloc[i]['Type'])
    Promoter_Name_list.append(Dataset.iloc[i]['Promoter'])
    
    Promoter_Address = ""
    Promoter_District = ""
    Promoter_Contact_Number = ""
    Promoter_Website = ""
    
    Project_Commencement_Date = ""
    Project_Completion_Date = ""
    
    Project_Tehsil = ""
    Project_Plot = ""
    Project_Street = ""
    Project_Town = ""
    
    Project_Number_of_Apartments = ""
    Project_Number_of_sanctioned_Apartments = ""
    
    Contractor_Name = ""
    Contractor_Email = ""
    Contractor_Address = ""
    Contractor_Contact_Number = ""
    
    Architect_Name = ""
    Architect_Email = ""
    Architect_Address = ""
    Architect_Contact_Number = ""
    
    Structural_Engineer_Name = ""
    Structural_Engineer_Email = ""
    Structural_Engineer_Address = ""
    Structural_Engineer_Contact_Number = ""
    
    
    tables = soup.find_all('table')
    for table in tables :
        
        #print(table)
        #print(type(table))
        sub_soup = BeautifulSoup(str(table), "html.parser")
        table_head = sub_soup.find('th')
        
        if(table_head):
            
            if("Address Details" in table_head.text):
                
                tds = sub_soup.find_all('td')
                for td in tds :
                   
                    #Promoter Stree/Locality
                    if("Street/ Locality" in td.text):
                        if(td.find_next_sibling('td')):
                            Promoter_Address= td.find_next_sibling('td').text
                        else :
                            pass
                    else:
                        pass
            
                    #Promoter District
                    if("District" in td.text):
                        if(td.find_next_sibling('td')):
                            Promoter_District = td.find_next_sibling('td').text
                        else:
                            pass
                    else:
                        pass
            
                   
                    
            elif("Organization Contact Details" in table_head.text):
                    
                    tds = sub_soup.find_all('td')
                    for td in tds :
                                      
                        #Promoter Contact
                        if("Office Number" in td.text):
                            if(td.find_next_sibling('td')):
                                Promoter_Contact_Number = td.find_next_sibling('td').text
                            else:
                                pass
                        else:
                            pass
        
                        #Promoter Website
                        if("Website URL" in td.text):
                            if(td.find_next_sibling('td')):
                                Promoter_Website = td.find_next_sibling('td').text
                            else:
                                pass
                        else:
                            pass
            
            elif("Land Details Of Registering Project" in table_head.text):
                
                    tds = sub_soup.find_all('td')
                    for td in tds :
                        
                        if("Number Of Apartments / Plots" in td.text):
                            if(td.find_next_sibling('td')):
                                Project_Number_of_Apartments = str(td.find_next_sibling('td').text)
                            else:
                                pass
                        else:
                            pass
                        
                        if("Sanctioned Number Of Apartments / Plots" in td.text):
                            if(td.find_next_sibling('td')):
                                Project_Number_of_sanctioned_Apartments = str(td.find_next_sibling('td').text)
                            else:
                                pass
                        else:
                            pass
                            
             
            elif("LOCATION OF PROJECT" in table_head.text):
                
                    #Project Location
                    tds = sub_soup.find_all('td')
                    for td in tds :
                                        
                        #Project Tehsil
                        if("Tehsil" in td.text):
                            if(td.find_next_sibling('td')):
                                Project_Tehsil = td.find_next_sibling('td').text
                            else:
                                pass
                        else:
                            pass
                    
                        #Project Town
                        if("Village/ Town/ City" in td.text):
                            if(td.find_next_sibling('td')):
                                Project_Town = td.find_next_sibling('td').text
                            else:
                                pass
                        else:
                            pass
                        
                        
                        #Project Plot
                        if("Plot / Khasra No" in td.text):
                            if(td.find_next_sibling('td')):
                                Project_Plot = td.find_next_sibling('td').text
                            else:
                                pass
                        else:
                            pass
                        
                        #Project Street
                        if("Street/ Locality" in td.text):
                            if(td.find_next_sibling('td')):
                                Project_Street = td.find_next_sibling('td').text
                            else:
                                pass
                        else :
                            pass
                        
                        
        else:
            pass
    
    
    tds = soup.find_all("td")
    for td in tds :
        
        if("Actual Commencement Date" in td.next):
            if(td.find_next_sibling('td')):
                Project_Commencement_Date = td.find_next_sibling('td').text
            else:
                pass
        else :
            pass
        
        if("Estimated Finish Date" in td.next):
            if(td.find_next_sibling('td')):
                Project_Completion_Date = td.find_next_sibling('td').text
            else:
                pass
        else:
            pass

        
    #Check if Professional Details is present    
    h3_header = soup.find_all("h3",class_="TableHeading")
    
    #print("-"*50)
    print("Row : "+str(i))
    for h3 in h3_header:
        if("CONTRACTOR" in h3.text):
            
            table = h3.findNext("table")
            table_header = table.find_all("th")
            table_value = table.find_all("td")
            #print(str(table_header))
            #print(str(table_value))
            for i in range(0,len(table_header)):
                if(table_header[i].text == "E-mail Address"):
                    Contractor_Email = table_value[i].text
                elif(table_header[i].text == "Name"):
                    Contractor_Name = table_value[i].text
                elif(table_header[i].text == "Contact Address"):
                    Contractor_Address = table_value[i].text
                elif(table_header[i].text == "Contact Number"):
                    Contractor_Contact_Number = table_value[i].text
            
            
            
        
        elif("ARCHITECT" in h3.text):
            
            table = h3.findNext("table")
            table_header = table.find_all("th")
            table_value = table.find_all("td")
            #print(str(table_header))
            #print(str(table_value))
            for i in range(0,len(table_header)):
                if(table_header[i].text == "E-mail Address"):
                    Architect_Email = table_value[i].text
                elif(table_header[i].text == "Name"):
                    Architect_Name = table_value[i].text
                elif(table_header[i].text == "Contact Address"):
                    Architect_Address = table_value[i].text
                elif(table_header[i].text == "Contact Number"):
                    Architect_Contact_Number = table_value[i].text
            
        elif("STRUCTURAL ENGINEER" in h3.text):
            
            table = h3.findNext("table")
            table_header = table.find_all("th")
            table_value = table.find_all("td")
            #print(str(table_header))
            #print(str(table_value))
            for i in range(0,len(table_header)):
                if(table_header[i].text == "E-mail Address"):
                    Structural_Engineer_Email = table_value[i].text
                elif(table_header[i].text == "Name"):
                    Structural_Engineer_Name = table_value[i].text
                elif(table_header[i].text == "Contact Address"):
                    Structural_Engineer_Address = table_value[i].text
                elif(table_header[i].text == "Contact Number"):
                    Structural_Engineer_Contact_Number = table_value[i].text
            
    
    
    #print(Contractor_Email)
    #print(Contractor_Name)
    #print(Contractor_Address)
    #print(Contractor_Contact_Number)
    
    Project_Contractor_Email_Address_list.append(Contractor_Email)
    Project_Contractor_Name_list.append(Contractor_Name)
    Project_Contractor_Contact_Address_list.append(Contractor_Address)
    Project_Contractor_Contact_Number_list.append(Contractor_Contact_Number)
    
    #print(Architect_Email)
    #print(Architect_Name)
    #print(Architect_Address)
    #print(Architect_Contact_Number)
    
    Project_Architect_Email_Address_list.append(Architect_Email)
    Project_Architect_Name_list.append(Architect_Name)
    Project_Architect_Contact_Address_list.append(Architect_Address)
    Project_Architect_Contact_Number_list.append(Architect_Contact_Number)
    
    #print(Structural_Engineer_Email)
    #print(Structural_Engineer_Name)
    #print(Structural_Engineer_Address)
    #print(Structural_Engineer_Contact_Number)
    
    Project_Structural_Engineer_Name_list.append(Structural_Engineer_Name)
    Project_Structural_Engineer_Email_Address_list.append(Structural_Engineer_Email)
    Project_Structural_Engineer_Contact_Address_list.append(Structural_Engineer_Address)
    Project_Structural_Engineer_Contact_Number_list.append(Structural_Engineer_Contact_Number)
            
    #print("-"*50)
    #print("\n")
    #print("\n")
    #print("\n")
    #Extract HTML between Project Professional Details Header and Next Sibling Header
    
            
            
     
    #print(Promoter_Address)
    #print(Promoter_District)
    #print(Promoter_Contact_Number)
    #print(Promoter_Website)
    
    #print(Project_Commencement_Date) 
    #print(Project_Completion_Date)
    
    #print(Project_Tehsil) 
    #print(Project_Plot )
    #print(Project_Street)
    #print(Project_Town)
    Promoter_Address_list.append(Promoter_Address.strip())
    Promoter_District_list.append(Promoter_District.strip())
    Promoter_Contact_Number_list.append(Promoter_Contact_Number.strip())
    Promoter_Website_list.append(Promoter_Website.strip())
    
    Project_Commencement_Date_list.append(Project_Commencement_Date.strip())
    Project_Completion_Date_list.append(Project_Completion_Date.strip())
    
    Project_Tehsil_list.append(Project_Tehsil.strip())
    Project_Plot_list.append(Project_Plot.strip())
    Project_Street_list.append(Project_Street.strip())
    Project_Town_list.append(Project_Town.strip())
    
    Project_Number_of_Apartments_list.append(Project_Number_of_Apartments)
    Project_Number_of_Sanctioned_Apartment_list.append(Project_Number_of_sanctioned_Apartments)
    
    driver.quit()
    
    
```
