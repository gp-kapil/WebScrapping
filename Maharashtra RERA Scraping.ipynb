```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from selenium import webdriver
import time
from selenium.webdriver.chrome.options import Options
```

```python
SerialNo = []
ProjectName = []
PromoterName = []
LastModifiedDate = []
ViewDetails = []
PromoterContactDetails = []
PromoterWebsite = []
ProjectStatus = []
ProjectDateOfCompletion = []
ProjectType = []
ProjectTotalBuildingCount = []
ProjectLocality = []
```

```python
def Scrap_Project_Details(URL):
    
    P_Contact_Details = ""
    P_Website = ""
    Pr_Status = ""
    Pr_Date_Of_Completion = ""
    Pr_Type = ""
    Pr_Total_Building_Count = ""
    Pr_Locality = ""
        
    print(URL)
    driver_f = webdriver.Firefox()
    driver_f.get(URL)
    soup = BeautifulSoup(driver_f.page_source, "html.parser")
    
    divs = soup.find_all('div',class_="col-md-3 col-sm-3")
    
    for div in divs:
        
        if(div.find('label')):

                label = div.label.text
                
                if(label.strip() == "Office Number"):
                    P_Contact_Details = div.find_next_sibling("div").get_text().strip()
                elif(label.strip() == "Website URL"):
                    P_Website = div.find_next_sibling("div").get_text().strip()
                elif(label.strip() == "Project Status"):
                    Pr_Status = div.find_next_sibling("div").get_text().strip()
                elif(label.strip() == "Revised Proposed Date of Completion"):
                    Pr_Date_Of_Completion = div.find_next_sibling("div").get_text().strip()
                elif(label.strip() == "Project Type"):
                    Pr_Type = div.find_next_sibling("div").get_text().strip()
                elif(label.strip() == "Locality"):
                    Pr_Locality = div.find_next_sibling("div").get_text().strip()
                elif(label.strip() == "Total Building Count"):
                    Pr_Total_Building_Count = div.find_next_sibling("div").get_text().strip()
          
    PromoterContactDetails.append(P_Contact_Details)
    PromoterWebsite.append(P_Website)
    ProjectStatus.append(Pr_Status)
    ProjectDateOfCompletion.append(Pr_Date_Of_Completion)
    ProjectType.append(Pr_Type)
    ProjectTotalBuildingCount.append(Pr_Total_Building_Count)
    ProjectLocality.append(Pr_Locality)
        
    driver_f.quit()                
```

```python
def Scrap_Project_URL():
    next_button_id = driver.find_element_by_id("btnNext")
    while(True):
        
        if(next_button_id.is_enabled):
            soup = BeautifulSoup(driver.page_source, "html.parser")
            table_rows = soup.find_all('tr',class_="grid-row")
            #counter = 0
            for row in table_rows:
                #counter +=1
                #print(counter)
                table_div = row.find_all('td',class_="grid-cell")
            
                for td in table_div:
                    if(td['data-name'] == 'Srno'):
                        SerialNo.append(td.text)
                    elif(td['data-name']=='Project'):
                        ProjectName.append(td.text)
                    elif(td['data-name'] == 'Name'):
                        PromoterName.append(td.text)
                    elif(td['data-name'] == 'lastModifiedDate'):
                        LastModifiedDate.append(td.text)
            
                anchors = row.find_all('a')
                for anchor in anchors:
                    if(anchor.text.strip() == 'View'):
                        ViewDetails.append('https://maharerait.mahaonline.gov.in'+anchor['href'])
                        Scrap_Project_Details('https://maharerait.mahaonline.gov.in'+anchor['href'])
                        break
            
            next_button_id.click()
            time.sleep(4)
            next_button_id = driver.find_element_by_id("btnNext")
        else:
            print("Scraping Done")
            return
        
```
